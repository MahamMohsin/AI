```python
class student:
    def __init__(self,name,rollnum): #constructor
        self.name=name
        self.rollnum=rollnum

    def display(self):  #display
        print("Name: ",self.name)
        print("Roll_number: ",self.rollnum)

class exam(student):
    def __init__ (self,name,rollnum,marks):
        super() .__init__(name,rollnum)
        self.marks=marks

    def average(self):
        return sum(self.marks) / len(self.marks)

    def display_result(self):
        self.display()
        print("Avg marks: ",self.average())

s1=exam("Sidra",123,[70,80,90])
s2=exam("Ali",111,[80,80,80])

s1.display_result()
s2.display_result()
```

```python
class Bank:
    #__var → private
    #__private, access only via methods
    def __init__(self,accnum,balance):
        self.__accnum=accnum
        self.__balance=balance

    def deposit(self,amount):
        self.__balance+=amount

    def withdraw(self,amount):
        if amount>=self.__balance:
            print("Insufficient balance!")
        else:
            self.__balance-=amount

    def display(self):
        print("Account Number: ",self.__accnum)
        print("Balance: ",self.__balance)


obj=Bank(12345,10000)
obj.deposit(2000)
obj.display()
obj.withdraw(1000)
obj.display()
obj.withdraw(11000)
```

```python
#method overiding 
#same function name, different behavior
class employee:
    def __init__(self,name,sal):
        self.name=name
        self.sal=sal

    def get_details(self):
        print("Name: ",self.name)
        print("Salary: ",self.sal)

class manager(employee):
    def __init__(self,name,sal,bonus):
        super().__init__(name,sal)
        self.bonus=bonus

    def get_details(self):#ovride
        print("Manager details: ")
        print("Name: ",self.name)
        print("Salary: ",self.sal)
        print("bonus: ",self.bonus)
        print("Total Salary: ",self.sal+self.bonus)

class developer(employee):
    def __init__(self,name,sal,project):
        super().__init__(name,sal)
        self.project=project

    def get_details(self):
        print("Developer details: ")
        print("Name: ",self.name)
        print("Salary: ",self.sal)
        print("Project: ",self.project)

m1=manager("Ali",25000,10000)
d1=developer("Sara",30000,"React landing page")

m1.get_details()
d1.get_details()
```
