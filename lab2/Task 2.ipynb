```python
#Base class
class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number

    def display(self):
        print("Student Details:")
        print("Name:", self.name)
        print("Roll Number:", self.roll_number)


#Derived class
class Exam(Student):
    def __init__(self, name, roll_number, marks):
        super().__init__(name, roll_number)   #calling parent constructor
        self.marks = marks

    def average(self):
        return sum(self.marks) / len(self.marks)

    def display_result(self):
        self.display()   # calling display method of Student
        print("Average Marks:", self.average())


#creating objects
student1 = Exam("Ali", 101, [80, 90, 85])
student2 = Exam("Sara", 102, [75, 85, 95])

#displaying results
student1.display_result()
student2.display_result()
```
