```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def get_details(self):
        print("Name:", self.name)
        print("Salary:", self.salary)


class Manager(Employee):
    def __init__(self, name, salary, bonus):
        super().__init__(name, salary)
        self.bonus = bonus

    def get_details(self):
        print("Manager Details:")
        print("Name:", self.name)
        print("Salary:", self.salary)
        print("Bonus:", self.bonus)
        print("Total Compensation:", self.salary + self.bonus)


class Developer(Employee):
    def __init__(self, name, salary, project):
        super().__init__(name, salary)
        self.project = project

    def get_details(self):
        print("Developer Details:")
        print("Name:", self.name)
        print("Salary:", self.salary)
        print("Project:", self.project)

#Manager input
m_name = input("Manager Name: ")
m_salary = int(input("Salary: "))
m_bonus = int(input("Bonus: "))

manager = Manager(m_name, m_salary, m_bonus)

print()

#Developer input
d_name = input("Developer Name: ")
d_salary = int(input("Salary: "))
d_project = input("Project: ")

developer = Developer(d_name, d_salary, d_project)

print("\nOutput:")
manager.get_details()
print()
developer.get_details()
```
