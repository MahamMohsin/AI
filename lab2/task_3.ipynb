Q) Write a Python program to create a class named BankAccount that demonstrates private members.
1. The class should have the following private data members:
*   __account_number
*   __balance



2. Create a constructor to initialize the account number and balance.

3. Create the following public methods:

*   deposit(amount) → adds amount to balance

*   withdraw(amount) → subtracts amount from balance if sufficient funds are available, otherwise prints "Insufficient balance"

*   display() → prints the account number and balance

4. Create an object of the class, perform some deposits and withdrawals, and display the final
account details.

 **Sample Input**

Account Number: 12345

Initial Balance: 5000

Deposit: 2000

Withdraw: 3000

 **Sample Output**

Account Number: 12345

Balance: 4000

```python
class BankAccount:
  def __init__(self,accnum,bal):
    self.__account_number=accnum
    self.__balance=bal

  def deposit(self,amount):
    self.__balance+=amount

  def withdraw(self,amount):
    if self.__balance>=amount:
       self.__balance-=amount
    else:
      print("Insufficient Balance!!")

  def display(self):
    print("Account Number: ",self.__account_number)
    print("Balance: ",self.__balance)

accnum=int(input("Enter Account Number: "))
bal=float(input("Enter Initial Balance: "))

BA=BankAccount(accnum,bal)

amt=float(input("Enter amount to deposit: "))
BA.deposit(amt)

amt2=float(input("Enter amount to withdraw: "))
BA.withdraw(amt2)

BA.display()
```
