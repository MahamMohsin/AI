```python
#TASK 4
t = (10, 5, 8, 3, 12)
print("Original tuple:", t)

lst = list(t)
lst.append(7)
t = tuple(lst)
print("After adding an element:", t)

lst = list(t)
lst.insert(2, 15)  
t = tuple(lst)
print("After inserting an element:", t)

lst = list(t)
lst.remove(8)     
t = tuple(lst)
print("After removing an element:", t)

lst = list(t)
lst.sort()
t = tuple(lst)
print("After sorting:", t)

lst = list(t)
del lst[3]         
t = tuple(lst)
print("After deleting an element by index:", t)
```

```python
#TASK 5
s = {10, 5, 8, 3, 12}
print("Original set:", s)

s.add(7)
print("After adding one element:", s)

s.update([15, 2, 9])
print("After adding multiple elements:", s)

s.remove(8) 
print("After removing an element:", s)

sorted_set = sorted(s)
print("Set in ascending order:", sorted_set)

s.clear()
print("After clearing the set:", s)
```

```python
#TASK 6
students = {
    "Alice": 85,
    "Bob": 78,
    "Charlie": 92
}

print("Original dictionary:", students)

students["David"] = 88
print("After adding a new student:", students)

students["Bob"] = 82
print("After updating marks:", students)

del students["Alice"]
print("After deleting a student:", students)

print("Student Names:")
for name in students.keys():
    print(name)

print("Student Marks:")
for marks in students.values():
    print(marks)
```

```python
#TASK 7
marks = int(input("Enter student's marks (0–100): "))

# Check grade
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Grade: F")
```

```python
#TASK 9
def add_numbers(a, b=5):
    return a + b

print("Sum with one argument:", add_numbers(10))
print("Sum with two arguments:", add_numbers(10, 3))

def greet_user(name="Guest"):
    print("Hello,", name)

greet_user()
greet_user("Alice")

def power(num, exp=2):
    return num ** exp

print("Square of 4:", power(4))
print("2 raised to power 3:", power(2, 3))
```

```python
#TASK 10
def simple_interest(principal, time, rate=5):
    interest = (principal * time * rate) / 100
    return interest

print("Simple Interest:", simple_interest(1000, 2))
print("Simple Interest with custom rate:", simple_interest(1000, 2, 8))
```
