```python
#TASK 1
grid=[['S',1,'#',2,3],
     [2,'#',2,'#',1],
     [3,2,1,1,2],
      ['#',2,'#',1,3],
      [2,1,2,2,'G']]

#set start and end coords
start=(0,0)
goal=(4,4)

from collections import deque

#bfs path from s to g - gsg 
def bfs(grid,start,goal):
    #INITIALIZE R,C,QUEUE,VIISTED,DIRECTIONS
    
    #assign rows and columns
    rows=len(grid)
    cols=len(grid[0])
    #coordinate and path
    queue=deque([(start, [start])])
    visited=set()
    visited.add(start)
    
    directions=[(-1,0),(1,0),(0,-1),(0,1)]

    while queue:
        (x,y),path=queue.popleft()

        if (x,y) == goal:
            return path

        #check neighbours by first move then check boundary for adding in path
        for dx,dy in directions:
            nx,ny = x+dx, y+dy

            if 0<=nx<rows and 0<=ny<cols:
                if grid[nx][ny] != '#' and (nx,ny) not in visited:
                    visited.add((nx,ny))
                    queue.append(((nx,ny), path + [(nx,ny)]))

    return None

print("BFS PATH: ",bfs(grid,start,goal))
```

```python
#DFS - ITERATIVE
def dfs(grid, start, goal):
    rows = len(grid)
    cols = len(grid[0])
    stack = [(start, [start])]
    visited = set() #no need to add start after it

    directions = [(-1,0),(1,0),(0,-1),(0,1)]

    while stack:
        (x, y), path = stack.pop()

        if (x, y) == goal:
            return path

        if (x, y) not in visited:
            visited.add((x, y))

            for dx, dy in directions:
                nx, ny = x + dx, y + dy

                if 0 <= nx < rows and 0 <= ny < cols:
                    if grid[nx][ny] != '#' and (nx, ny) not in visited:
                        stack.append(((nx, ny), path + [(nx, ny)]))

    return None
print("DFS ITERATIVE PATH: ", dfs(grid,start,goal))
```

```python
#UNIFORM COST SEARCH - BASED ON COST 
#--same as dfs use import heapq and stack ki jaga pq and heappq.heappop(pq) 
#and push mein pq mein pq and agy ka bhi hoga

import heapq
grid=[['S',1,'#',2,3],
     [2,'#',2,'#',1],
     [3,2,1,1,2],
      ['#',2,'#',1,3],
      [2,1,2,2,'G']]

start=(0,0)
goal=(4,4)

def ucs(grid,start,goal):
    #initialize r,c,visited,directions,pq

    rows=len(grid)
    cols=len(grid[0])
    pq=[(0, start, [start])] #cost,node,path
    visited=set()
    directions = [(-1,0),(1,0),(0,-1),(0,1)]

    while pq:
        cost,(x,y),path=heapq.heappop(pq)

        if (x,y) == goal:
            return path,cost

        if (x,y) not in visited:
            visited.add((x,y))

        #now neighbours
        for dx,dy in directions:
            nx, ny = x + dx, y + dy

            #boundary check
            if 0 <= nx < rows and 0 <= ny < cols:
                if grid[nx][ny] != '#' and (nx,ny) not in visited:
                    newcost=cost + (0 if grid[nx][ny] == 'G' else grid[nx][ny])
                    heapq.heappush(pq,(newcost,(nx,ny), path + [(nx,ny)]))
    return None

path, cost = ucs(grid, start, goal)
print("UCS Path:", path)
print("Total Cost:", cost)
```

```python
#DLmitedSearch
def dls_iterative(grid, start, goal, limit):
    rows, cols = len(grid), len(grid[0])
    stack = [(start, [start], 0)]  # node, path, depth
    visited = set()

    directions = [(-1,0),(1,0),(0,-1),(0,1)]

    while stack:
        (x, y), path, depth = stack.pop()

        if (x, y) == goal:
            return path

        if depth < limit and (x, y) not in visited:
            visited.add((x, y))

            for dx, dy in directions:
                nx, ny = x + dx, y + dy

                if 0 <= nx < rows and 0 <= ny < cols:
                    if grid[nx][ny] != '#':
                        stack.append(((nx, ny), path + [(nx, ny)], depth + 1))

    return None
```

```python
import heapq
#GREEDY BEST FIRST SEARCH
#since grid mein hee heuristic values same as ucs
grid = [
    ['S', 4, '#', 2, 0],
    [3, '#', 3, '#', 1],
    [2, 2, 2, 2, 1],
    ['#', 3, '#', 1, 2],
    [3, 2, 3, 2, 'G']
]
start=(0,0)
goal=(4,4)
def greedy(grid, start, goal):
    rows, cols = len(grid), len(grid[0])
    pq = [(0, start, [start])]  # (heuristic, node, path)
    visited = set()

    directions = [(-1,0),(1,0),(0,-1),(0,1)]

    while pq:
        h, (x, y), path = heapq.heappop(pq)

        if (x, y) == goal:
            return path

        if (x, y) in visited:
            continue

        visited.add((x, y))

        for dx, dy in directions:
            nx, ny = x + dx, y + dy

            if 0 <= nx < rows and 0 <= ny < cols:
                if grid[nx][ny] != '#':
                    h_val = 0 if grid[nx][ny] in ['S','G'] else int(grid[nx][ny])
                    heapq.heappush(pq, (h_val, (nx, ny), path + [(nx, ny)]))

    return None

print("Greedy Path:", greedy(grid, start, goal))
```

```python
#A-STAR ALGO
#same as dfs krna and use heapq like ucs and gbfs
import heapq
grid = [
    ['S', 4, '#', 2, 0],
    [3, '#', 3, '#', 1],
    [2, 2, 2, 2, 1],
    ['#', 3, '#', 1, 2],
    [3, 2, 3, 2, 'G']
]

start=(0,0)
goal=(4,4)

def astar(grid,start,goal):
    #rows,cols,visited,dirs,pq initialized
    rows=len(grid)
    cols=len(grid[0])
    visited=set()
    directions=[(-1,0),(1,0),(0,-1),(0,1)]
    pq=[(0,0,start,[start])] #f,g,node,path

    while pq:
        #g is cost, f is func f=g+h
        f,g,(x,y),path=heapq.heappop(pq)

        if (x,y) == goal:
            return path,g

        if (x,y) not in visited:
            visited.add((x,y))

            for dy,dx in directions:
                nx,ny=x+dx,y+dy

                if 0<=nx<rows and 0<=ny<cols:
                    if grid[nx][ny] != '#' and (nx,ny) not in visited:
                        gnew=g+1 #cost to move cell is 1
                        h=0 if grid[nx][ny] == 'G' else int(grid[nx][ny])
                        fnew=gnew+h

                        heapq.heappush(pq,(fnew,gnew,(nx,ny),path+[(nx,ny)]))
    return None

path, cost = astar(grid, start, goal)
print("A* Path:", path)
print("Cost:", cost)  
```
