```python
#TASK 1
grid =[
    ['S', 2, 0, 0, 1],
    [0, 'X', 1, 2, 0],
    [0, 2, 0, 'X', 0],
    [0, 0, 1, 0, 2]
]

rows=len(grid)
cols=len(grid[0])

#storing start coordinates
for i in range(rows):
    for j in range(cols):
        if grid[i][j]=='S':
            x,y=i,j

#initialization
visited=set() #no dups,fast look up
visited.add((x,y))
total_score=0

#posssible dirctions U,D,L,R
directions=[(-1,0),(1,0),(0,-1),(0,1)]

path=[(x,y)]

#total possible moves=8
moves=1
#for _ in range(moves) ----- moves=8
while moves <= 8:
    best= None
    best_val=-1

    for dx,dy in directions:
        #checking neighbours
        nx,ny= x+dx, y+dy

        #checking boundary conditions
        if 0<=nx<rows and 0<=ny<cols:
            if grid[nx][ny] != 'X' and (nx,ny) not in visited:
                #val mein stored if int value else 0
                val=grid[nx][ny] if isinstance(grid[nx][ny],int) else 0
                if val>best_val: #checking best_value_score
                    best_val=val
                    best=(nx,ny) #storing best move

    if best is None:
        break

    x,y=best #when best received store in x,y
    visited.add((x,y)) #add to visited
    total_score+=best_val
    path.append((x,y)) #adding in path
    moves+=1 #we have to keep incrementing else acts as infinite loop

print("Path: ",path)
print("Total score: ",total_score)
    
```

```python
heuristics = {"T0":5, "T1":3, "T2":4, "T3":2, "T4":1, "T5":2, "T6":0}

graph = {
    "T0": ["T1","T2"],
    "T1": ["T3","T4"],
    "T2": ["T5"],
    "T3": ["T6"],
    "T4": ["T6"],
    "T5": ["T6"],
    "T6": []
}

completed = set()
order = []

while len(completed) < len(graph):
    available = []

    for node in graph:
        if node not in completed:
            # check dependencies
            deps_satisfied = True
            for parent in graph:
                if node in graph[parent] and parent not in completed:
                    deps_satisfied = False

            if deps_satisfied:
                available.append(node)

    # pick lowest heuristic
    best = min(available, key=lambda x: heuristics[x])
    order.append(best)
    completed.add(best)

print("Order:", order)
```

```python
#DLS
def dls_iterative(grid, start, goal, limit):
    rows, cols = len(grid), len(grid[0])
    stack = [(start, [start], 0)]  # node, path, depth
    visited = set()

    directions = [(-1,0),(1,0),(0,-1),(0,1)]

    while stack:
        (x, y), path, depth = stack.pop()

        if (x, y) == goal:
            return path

        if depth < limit and (x, y) not in visited:
            visited.add((x, y))

            for dx, dy in directions:
                nx, ny = x + dx, y + dy

                if 0 <= nx < rows and 0 <= ny < cols:
                    if grid[nx][ny] != '#':
                        stack.append(((nx, ny), path + [(nx, ny)], depth + 1))

    return None
```
