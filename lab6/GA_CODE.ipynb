# 8-Queens Problem solved using Genetic Algorithm
# Each chromosome represents one board configuration
# Index = column, Value = row

```python
import random  # Import random module for generating random numbers
```

```python

# Number of queens / size of the board
N = 8
```

```python
# ---------------------------
# 1️. Fitness Function
# ---------------------------
# Computes how "good" a chromosome is.
# Maximum fitness = 28 (all pairs of queens are non-attacking)
def fitness(chromosome):
    conflicts = 0  # Counter for attacking pairs

    # Compare all pairs of queens
    for i in range(N):
        for j in range(i + 1, N):
            # Row conflict: same row
            if chromosome[i] == chromosome[j]:
                conflicts += 1
            # Diagonal conflict: difference in rows == difference in columns
            if abs(chromosome[i] - chromosome[j]) == abs(i - j):
                conflicts += 1

    # Fitness = max non-attacking pairs (28) - conflicts
    return 28 - conflicts
```

```python
# ---------------------------
# 2️. Create Random Chromosome
# ---------------------------
# Generates a random board configuration
def create_chromosome():
    # Random row (1-8) for each column
    return [random.randint(1, N) for _ in range(N)]
```

```python
# ---------------------------
# 3️. Selection
# ---------------------------
# Selects the top 2 chromosomes (best fitness) as parents
def selection(population):
    # Sort population in descending order of fitness
    population = sorted(population, key=lambda x: fitness(x), reverse=True)
    # Return the top 2 chromosomes
    return population[:2]
```

```python
# ---------------------------
# 4️. Crossover
# ---------------------------
# Combines two parent chromosomes to create a child
def crossover(parent1, parent2):
    # Choose a random crossover point (avoid first and last index)
    point = random.randint(1, N - 2)
    # Child takes first part from parent1, second part from parent2
    child = parent1[:point] + parent2[point:]
    return child
```

```python
# ---------------------------
# 5️. Mutation
# ---------------------------
# Introduces random variation in a chromosome
def mutation(chromosome):
    # Pick a random index (column) to change
    index = random.randint(0, N - 1)
    # Assign a new random row (1-8) for that column
    chromosome[index] = random.randint(1, N)
    return chromosome
```

```python
# ---------------------------
# 6️. Genetic Algorithm
# ---------------------------
def genetic_algorithm():
    population_size = 100  # Number of chromosomes per generation
    generations = 1000     # Maximum number of generations

    # Step 1: Initialize population with random chromosomes
    population = [create_chromosome() for _ in range(population_size)]

    # Step 2: Evolution loop
    for generation in range(generations):
        # Check if any chromosome is a perfect solution
        for individual in population:
            if fitness(individual) == 28:  # Maximum fitness = 28
                print("Solution found:", individual)
                return individual  # Stop and return solution

        # Step 3: Select best parents
        parent1, parent2 = selection(population)

        # Step 4: Generate new population
        new_population = []
        for _ in range(population_size):
            # Crossover to produce a child
            child = crossover(parent1, parent2)

            # Apply mutation with 10% chance
            if random.random() < 0.1:
                child = mutation(child)

            # Add child to new population
            new_population.append(child)

        # Step 5: Replace old population with new generation
        population = new_population

    # If no solution found after all generations
    print("No perfect solution found.")
    return None

# ---------------------------
# 7️⃣ Run the Genetic Algorithm
# ---------------------------
solution = genetic_algorithm()
```
