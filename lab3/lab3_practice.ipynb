```python
#task 1 says no memory means reflex based agent model

class lightAgent:
    def __init__(self,rooms):
        self.rooms=rooms

    def act(self):
        for room in self.rooms:
            if self.rooms[room]=="occupied":
                print(f"{room}: LIGHT ON")

            else:
                print(f"{room}: LIGHT OFF")

#env
rooms={}
n=int(input("Enter room nums: "))

for i in range(n):
    status=input(f"Enter status of Room{i+1}: ")
    rooms[f"Room{i+1}"]=status

agent=lightAgent(rooms)
#for continuous system
while True:
    agent.act()

    change=input("Update room status or type exit: ")
    if change.lower()=="exit":
        break

    room_name, state=change.split()
    rooms[room_name]=state
```

```python
import random

class environment:
    def __init__(self):
        self.shelves=["A", "B", "C", "D", "E"]
        self.items={shelf: random.choice([True,False]) for shelf in self.shelves} 
        #true means item exists 

    def get_percept(self,shelf):
        return self.items[shelf] #returns value corresponding to key like key->value

class robotagent:
    def __init__(self):
        self.checkedshelves=[]
        self.collecteditems=[]

    def act(self,shelf,percept):
        if shelf in self.checkedshelves:
            return "Skip alr checked"

        self.checkedshelves.append(shelf)

        if percept:
            self.collecteditems.append(shelf)
            return f"Picked item from shelf {shelf}"
        else:
            return f"No item at shelf {shelf}"

def run():
    env=environment()
    agent=robotagent()

    print("Initial Shelf State: ",env.items)

    for shelf in env.shelves:
        percept=env.get_percept(shelf)
        action=agent.act(shelf,percept)

        print(f"Shelf {shelf}-> Percept: {percept}, Action: {action}")

    print ("Collected items: ",agent.collecteditems)

run()
```

```python
import random

class environment:
    def get_percept(self):
        return {
            "battery": random.randint(20,100), # % of battery remaining
            "weather": random.choice(["clear","windy","rainy","stormy"]),
            "distance": random.randint(1,20),
            "obstacles": random.randint(0,5)
        }


class droneAgent:
    def utility(self,action,percept):
        battery=percept["battery"]
        weather=percept["weather"]
        distance=percept["distance"]
        obstacles=percept["obstacles"]

        if action == "fly faster":
            return 10 - (100 - battery)/10 - obstacles*2 - (5 if weather != "Clear" else 0)
            #basically 10 as a base score taken

        elif action == "fly slower":
            return 6 + battery/20 - obstacles

        elif action == "alternative route":
            return 8 - distance/2 + (3 if obstacles > 2 else 0)


    def act(self,percept):
        actions=["fly faster","fly slower","alternative route"]

        utilities = {action: self.utility(action,percept) for action in actions}

        bestaction=max(utilities,key=utilities.get)
        #key=utilities.get → tells Python:
        #👉 “Compare based on values (50, 30, 40...) not keys
        return bestaction,utilities


def run():
    env=environment()
    agent=droneAgent()

    percept=env.get_percept()
    action,utilities=agent.act(percept)
    print("Percept:", percept)
    print("Utilities:", utilities)
    print("Chosen Action:", action)

run()
```
