```python
class LightControlAgent:
    def __init__(self, rooms):
        self.rooms = rooms

    def act(self):
        for room in self.rooms:
            if self.rooms[room] == "Occupied":
                print(f"{room}: Light ON")
            else:
                print(f"{room}: Light OFF")


#Environment
rooms = {
    "Room1": "Occupied",
    "Room2": "Empty",
    "Room3": "Occupied"
}

agent = LightControlAgent(rooms)

#Continuous Operation
while True:
    agent.act()
    
    change = input("Update room (e.g., Room1 Occupied/Empty) or type 'exit': ")
    if change.lower() == "exit":
        break
    
    room_name, state = change.split()
    rooms[room_name] = state
```
