```python
import heapq

graph = {
    "Central Command": [("Shelter A", 4), ("Shelter B", 2)],
    "Shelter A": [("Street 1", 5), ("Street 2", 3)],
    "Shelter B": [("Street 2", 2)],
    "Street 1": [("Hospital Zone", 7)],
    "Street 2": [("Hospital Zone", 4), ("Street 3", 6)],
    "Street 3": [("Hospital Zone", 2)],
    "Hospital Zone": []
}

def uniform_cost_search(graph, start, goal):
    pq = [(0, start, [start])]   
    visited = set()
    expansion_order = []

    while pq:
        cost, node, path = heapq.heappop(pq)

        if node not in visited:
            visited.add(node)
            expansion_order.append(node)

            if node == goal:
                print("Node Expansion Order:")
                print(" -> ".join(expansion_order))

                print("\nOptimal Path:")
                print(" -> ".join(path))

                print("\nTotal Cost:", cost, "minutes")
                return

            for neighbor, edge_cost in graph[node]:
                heapq.heappush(pq, (cost + edge_cost, neighbor, path + [neighbor]))

uniform_cost_search(graph, "Central Command", "Hospital Zone")
```
