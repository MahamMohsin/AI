```python
from collections import deque

graph = {
    "Control Room": ["Storage A", "Storage B"],
    "Storage A": ["Lab 1", "Lab 2"],
    "Storage B": ["Server Room"],
    "Lab 1": ["Debris 1"],
    "Lab 2": ["Debris 2"],
    "Debris 1": ["Maintenance Corridor"],
    "Debris 2": ["Maintenance Corridor"],
    "Server Room": ["Charging Bay"],
    "Charging Bay": ["Maintenance Corridor"],
    "Maintenance Corridor": ["Survivor Room"],
    "Survivor Room": []
}

start = "Control Room"
goal = "Survivor Room"

#task 1a: BFS - Breadth First Search 
def bfs(graph, start, goal):
    queue = deque([(start, [start])])
    visited = set()
    expansion_order = []

    while queue:
        node, path = queue.popleft()

        if node not in visited:
            visited.add(node)
            expansion_order.append(node)

            if node == goal:
                print("\nBFS Expansion Order:")
                print(" -> ".join(expansion_order))
                print("BFS Path to Goal:")
                print(" -> ".join(path))
                return

            for neighbor in graph[node]:
                queue.append((neighbor, path + [neighbor]))

#task 2b: DFS Recursive
def dfs_recursive(graph, node, goal, visited=None, path=None, expansion=None):
    if visited is None:
        visited = set()
        path = []
        expansion = []

    visited.add(node)
    path.append(node)
    expansion.append(node)

    if node == goal:
        print("\nDFS Recursive Expansion Order:")
        print(" -> ".join(expansion))
        print("DFS Recursive Path to Goal:")
        print(" -> ".join(path))
        return True

    for neighbor in graph[node]:
        if neighbor not in visited:
            if dfs_recursive(graph, neighbor, goal, visited, path, expansion):
                return True

    path.pop()
    return False
    
#task 3c: DFS Iterative 
def dfs_iterative(graph, start, goal):
    stack = [(start, [start])]
    visited = set()
    expansion_order = []

    while stack:
        node, path = stack.pop()

        if node not in visited:
            visited.add(node)
            expansion_order.append(node)

            if node == goal:
                print("\nDFS Iterative Expansion Order:")
                print(" -> ".join(expansion_order))
                print("DFS Iterative Path to Goal:")
                print(" -> ".join(path))
                return

            for neighbor in reversed(graph[node]):
                stack.append((neighbor, path + [neighbor]))

bfs(graph, start, goal)
dfs_recursive(graph, start, goal)
dfs_iterative(graph, start, goal)
```
