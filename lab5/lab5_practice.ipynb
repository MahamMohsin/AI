```python
def greedybfs(graph,heuristic,start,goal):
    frontier=[(start,heuristic[start])] #node,h
    visited= set()
    came_from={start: None}

    print("Expansion order: ")

    while frontier: 
        frontier.sort(key=lambda x: x[1]) #sort by h(n)
        node, _=frontier.pop(0)

        if node in visited:
            continue

        print(node,end="")
        visited.add(node)

        if node==goal:
            path=[]
            while node:
                path.append(node)
                node=came_from[node]

            path.reverse()
            print("\nPath: ",path)
            return

        for neighbor in graph[node]:
            if neighbor not in visited:
                came_from[neighbor]=node
                frontier.append((neigbor,heuristic[neigbour])
```
