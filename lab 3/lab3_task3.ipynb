```python
class Drone:

    def __init__(self, battery, weather, distance, obstacles):
        self.battery = battery
        self.weather = weather
        self.distance = distance
        self.obstacles = obstacles

    def weather_risk(self):
        risk_levels = {
            "Clear": 1,
            "Windy": 3,
            "Rainy": 5,
            "Stormy": 8
        }
        return risk_levels[self.weather]

    def calculate_utility(self, action):

        risk = self.weather_risk()

        # Base values
        speed_score = 0
        battery_score = 0
        safety_score = 0

        if action == "Fly Faster":
            speed_score = 10
            battery_score = self.battery - (self.distance * 3)
            safety_score = 10 - (risk + 2)

        elif action == "Fly Slower":
            speed_score = 5
            battery_score = self.battery - (self.distance * 1.5)
            safety_score = 10 - (risk - 1)

        elif action == "Alternative Route":
            speed_score = 6
            battery_score = self.battery - (self.distance * 2)
            safety_score = 10 - (risk - self.obstacles)

        # Normalize negative scores
        battery_score = max(battery_score, 0)
        safety_score = max(safety_score, 0)

        utility = (
            speed_score * 0.4 +
            battery_score * 0.3 +
            safety_score * 0.3
        )

        return utility

    def choose_best_action(self):

        actions = ["Fly Faster", "Fly Slower", "Alternative Route"]
        utilities = {}

        for action in actions:
            utilities[action] = self.calculate_utility(action)

        best_action = max(utilities, key=utilities.get)

        print("Utilities:", utilities)
        print("Best Action:", best_action)

        return best_action

drone = Drone(
    battery=60,
    weather="Rainy",
    distance=8,
    obstacles=3
)

drone.choose_best_action()
```
