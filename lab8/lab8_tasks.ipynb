```python
#TASK 1
import math

#initializing board
board = [[" " for _ in range(3)] for _ in range(3)]

HUMAN = "O"
AI = "X"

def print_board():
    for row in board:
        print(" | ".join(row))
    print()

#check winner
def check_winner():
    #rows & Columns
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]

    #diagonals
    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]

    return None

#check if board full
def is_full():
    for row in board:
        if " " in row:
            return False
    return True

#minimax Algorithm
def minimax(is_maximizing):
    winner = check_winner()

    if winner == AI:
        return 1
    elif winner == HUMAN:
        return -1
    elif is_full():
        return 0

    if is_maximizing:
        best_score = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = AI
                    score = minimax(False)
                    board[i][j] = " "
                    best_score = max(score, best_score)
        return best_score
    else:
        best_score = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = HUMAN
                    score = minimax(True)
                    board[i][j] = " "
                    best_score = min(score, best_score)
        return best_score

#finding best move for AI
def best_move():
    best_score = -math.inf
    move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == " ":
                board[i][j] = AI
                score = minimax(False)
                board[i][j] = " "

                if score > best_score:
                    best_score = score
                    move = (i, j)

    return move

#main game loop
def play_game():
    print("Tic-Tac-Toe Game Started!\n")
    print_board()

    while True:
        #human turn
        while True:
            try:
                row, col = map(int, input("Enter row and column (0-2): ").split())
                if board[row][col] == " ":
                    board[row][col] = HUMAN
                    break
                else:
                    print("Invalid move! Cell already occupied.")
            except:
                print("Invalid input! Try again.")

        print_board()

        if check_winner() == HUMAN:
            print("You win!")
            break
        if is_full():
            print("It's a draw!")
            break

        #AI turn
        i, j = best_move()
        board[i][j] = AI

        print("AI move:")
        print_board()

        if check_winner() == AI:
            print("AI wins!")
            break
        if is_full():
            print("It's a draw!")
            break

play_game()
```

```python
#TASK 2
import math

#Alpha-Beta function
def alpha_beta(coins, left, right, is_max, alpha, beta):
    #base case
    if left > right:
        return 0

    if is_max:
        max_eval = -math.inf

        #pick left
        pick_left = coins[left] + alpha_beta(coins, left + 1, right, False, alpha, beta)
        max_eval = max(max_eval, pick_left)
        alpha = max(alpha, max_eval)

        #prune
        if beta <= alpha:
            return max_eval

        #pick right
        pick_right = coins[right] + alpha_beta(coins, left, right - 1, False, alpha, beta)
        max_eval = max(max_eval, pick_right)
        alpha = max(alpha, max_eval)

        return max_eval

    else:
        min_eval = math.inf

        #min picks left
        pick_left = alpha_beta(coins, left + 1, right, True, alpha, beta)
        min_eval = min(min_eval, pick_left)
        beta = min(beta, min_eval)

        #prune
        if beta <= alpha:
            return min_eval

        #min picks right
        pick_right = alpha_beta(coins, left, right - 1, True, alpha, beta)
        min_eval = min(min_eval, pick_right)
        beta = min(beta, min_eval)

        return min_eval


#function to simulate the actual game
def play_game(coins):
    left = 0
    right = len(coins) - 1

    max_score = 0
    min_score = 0

    print("Initial Coins:", coins)

    while left <= right:
        #max's turn
        pick_left = coins[left] + alpha_beta(coins, left + 1, right, False, -math.inf, math.inf)
        pick_right = coins[right] + alpha_beta(coins, left, right - 1, False, -math.inf, math.inf)

        if pick_left >= pick_right:
            choice = coins[left]
            max_score += choice
            left += 1
        else:
            choice = coins[right]
            max_score += choice
            right -= 1

        print(f"Max picks {choice}, Remaining Coins: {coins[left:right+1]}")

        if left > right:
            break

        #min's turn (greedy: smallest end)
        if coins[left] <= coins[right]:
            choice = coins[left]
            min_score += choice
            left += 1
        else:
            choice = coins[right]
            min_score += choice
            right -= 1

        print(f"Min picks {choice}, Remaining Coins: {coins[left:right+1]}")

    #final result
    print(f"\nFinal Scores - Max: {max_score}, Min: {min_score}")

    if max_score > min_score:
        print("Winner: Max")
    elif min_score > max_score:
        print("Winner: Min")
    else:
        print("It's a Draw!")

coins = [3, 9, 1, 2, 7, 5]
play_game(coins)
```

```python
import chess
import math

#piece values
piece_values = {
    chess.PAWN: 100,
    chess.KNIGHT: 320,
    chess.BISHOP: 330,
    chess.ROOK: 500,
    chess.QUEEN: 900,
    chess.KING: 20000
}

#evaluation Function
def evaluate_board(board):
    if board.is_checkmate():
        if board.turn:
            return -99999
        else:
            return 99999
    if board.is_stalemate() or board.is_insufficient_material():
        return 0

    score = 0

    for piece_type in piece_values:
        score += len(board.pieces(piece_type, chess.WHITE)) * piece_values[piece_type]
        score -= len(board.pieces(piece_type, chess.BLACK)) * piece_values[piece_type]

    # Bonus for mobility (number of legal moves)
    score += len(list(board.legal_moves)) * 0.1 if board.turn else -len(list(board.legal_moves)) * 0.1

    return score


#alpha-Beta Minimax
def minimax(board, depth, alpha, beta, is_maximizing):
    if depth == 0 or board.is_game_over():
        return evaluate_board(board)

    if is_maximizing:
        max_eval = -math.inf
        for move in board.legal_moves:
            board.push(move)
            eval = minimax(board, depth - 1, alpha, beta, False)
            board.pop()
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break  # Pruning
        return max_eval
    else:
        min_eval = math.inf
        for move in board.legal_moves:
            board.push(move)
            eval = minimax(board, depth - 1, alpha, beta, True)
            board.pop()
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break  # Pruning
        return min_eval


#find Best Move
def get_best_move(board, depth):
    best_move = None
    best_value = -math.inf if board.turn else math.inf

    for move in board.legal_moves:
        board.push(move)
        value = minimax(board, depth - 1, -math.inf, math.inf, not board.turn)
        board.pop()

        if board.turn:  
            if value > best_value:
                best_value = value
                best_move = move
        else:  
            if value < best_value:
                best_value = value
                best_move = move

    return best_move


#play Game (Human vs AI)
def play_game():
    board = chess.Board()
    depth = 3 

    print("Chess Game Started!\n")

    while not board.is_game_over():
        print(board)
        print("\n")

        if board.turn:  #white (Human)
            move = input("Enter your move (e.g., e2e4): ")
            try:
                board.push_uci(move)
            except:
                print("Invalid move, try again!")
                continue
        else:  #black (AI)
            print("AI is thinking...")
            move = get_best_move(board, depth)
            board.push(move)
            print(f"AI plays: {move}")

    print("\nFinal Board:")
    print(board)
    print("\nResult:", board.result())

play_game()
```

```python
import math

VALID_WORDS = {
    "cat", "bat", "rat", "car", "bar", "tar", "art", "arc"
}

PENALTY = -5

#check if valid word
def is_valid(word):
    return word in VALID_WORDS

#evaluation function
def evaluate(word):
    if is_valid(word):
        return len(word)
    else:
        return PENALTY


#minimax
def minimax(letters, current_word, is_max):
    # Base case: no letters left
    if not letters:
        return evaluate(current_word)

    if is_max:
        best = -math.inf
        for i in range(len(letters)):
            new_letters = letters[:i] + letters[i+1:]
            new_word = current_word + letters[i]
            score = minimax(new_letters, new_word, False)
            best = max(best, score)
        return best
    else:
        best = math.inf
        for i in range(len(letters)):
            new_letters = letters[:i] + letters[i+1:]
            new_word = current_word + letters[i]
            score = minimax(new_letters, new_word, True)
            best = min(best, score)
        return best

def best_move(letters, current_word):
    best_score = -math.inf
    move = None

    for i in range(len(letters)):
        new_letters = letters[:i] + letters[i+1:]
        new_word = current_word + letters[i]

        score = minimax(new_letters, new_word, False)

        if score > best_score:
            best_score = score
            move = letters[i]

    return move

def play_game(letters):
    current_word = ""
    is_max_turn = True

    print("Initial Letters:", letters)

    while letters:
        if is_max_turn:
            letter = best_move(letters, current_word)
            print(f"Max picks: {letter}")
        else:
            letter = None
            best_score = math.inf

            for i in range(len(letters)):
                new_letters = letters[:i] + letters[i+1:]
                new_word = current_word + letters[i]
                score = minimax(new_letters, new_word, True)

                if score < best_score:
                    best_score = score
                    letter = letters[i]

            print(f"Min picks: {letter}")

        current_word += letter
        letters.remove(letter)

        print("Current word:", current_word)
        print("Remaining letters:", letters)
        print()

        is_max_turn = not is_max_turn

    final_score = evaluate(current_word)
    print("Final word:", current_word)

    if final_score > 0:
        print("Valid word! Score:", final_score)
    else:
        print("Invalid word! Penalty:", final_score)

letters = ['c', 'a', 't']
play_game(letters)
```
