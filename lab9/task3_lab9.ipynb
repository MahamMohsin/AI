```python
#prior probabilities
P_Flu = 0.3
P_Cold = 0.7

#conditional probabilities
P_Fever_given_Flu = 0.9
P_Fever_given_Cold = 0.5

P_Cough_given_Flu = 0.8
P_Cough_given_Cold = 0.6

P_Chills_given_Flu = 0.6
P_Chills_given_Cold = 0.4

P_Fatigue_given_Flu = 0.7

#task 1: Fever = Yes, Cough = Yes
flu = P_Fever_given_Flu * P_Cough_given_Flu * P_Flu
cold = P_Fever_given_Cold * P_Cough_given_Cold * P_Cold

total = flu + cold

print("Task 1:")
print("P(Flu | Fever, Cough) =", flu / total)
print("P(Cold | Fever, Cough) =", cold / total)

#task 2: Fever = Yes, Cough = Yes, Chills = Yes
flu = P_Fever_given_Flu * P_Cough_given_Flu * P_Chills_given_Flu * P_Flu
cold = P_Fever_given_Cold * P_Cough_given_Cold * P_Chills_given_Cold * P_Cold

total = flu + cold

print("\nTask 2:")
print("P(Flu | Fever, Cough, Chills) =", flu / total)
print("P(Cold | Fever, Cough, Chills) =", cold / total)

#task 3: P(Fatigue = Yes | Flu)
print("\nTask 3:")
print("P(Fatigue = Yes | Flu) =", P_Fatigue_given_Flu)
```
