```python
#CSP
#step1: import
from ortools.sat.python import cp_model
model=cp_model.CpModel()

#step2: making vars
#since 1-5 time slots either can be assigned
t1=model.NewIntVar(1,5,"T1_ML")
t2=model.NewIntVar(1,5,"T2_robotics")
t3=model.NewIntVar(1,5,"T3_NLP")
t4=model.NewIntVar(1,5,"T4_CV")
t5=model.NewIntVar(1,5,"T5_RL")

presentations= [t1,t2,t3,t4,t5]

#step3: assigning domains with constraints
#1. all diff constraints
model.AddAllDifferent(presentations)

#2. t1 and t3 not consecutive
#means neither t1 can come before or after t3
model.add(t1-t3 != 1)
model.add(t1-t3 != -1)

#3. t5 cannot be slot 1 or 2
#model.add_forbidden_assignments([presentations[4]], [(1,), (2,)])
model.add(t5 != 1)
model.add(t5 != 2)

#4. t1 must come before t3
model.add(t1<t3)

#5. t4 cannot be in slot 5
model.add(t4 != 5)

#6. slot 3 can only be given to t1 or t2
model.AddForbiddenAssignments([t3],[(3,)])
model.AddForbiddenAssignments([t4],[(3,)])
model.AddForbiddenAssignments([t5],[(3,)])

#7. t2 must be before t4
model.add(t2<t4)

# solve csp
solver=cp_model.CpSolver()
status=solver.Solve(model)

slot_times = {
    1: "9:00-10:00 AM",
    2: "10:00-11:00 AM",
    3: "11:00-12:00 PM",
    4: "2:00-3:00 PM",
    5: "3:00-4:00 PM"
}
if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
  print("solution found!")
  print(f"T1(ML) -> slot {solver.Value(t1)} | {slot_times[solver.Value(t1)]} ")
  print(f"T2(robotics) -> slot {solver.Value(t2)} | {slot_times[solver.Value(t2)]} ")
  print(f"T3(NLP) -> slot {solver.Value(t3)} | {slot_times[solver.Value(t3)]} ")
  print(f"T4(CV) -> slot {solver.Value(t4)}  | {slot_times[solver.Value(t4)]} ")
  print(f"T5(RL) -> slot {solver.Value(t5)}  | {slot_times[solver.Value(t5)]} ")
else:
  print ("NO SOLUTION!")
```
